```python
import numpy as np
import matplotlib.pyplot as plt

# a module made of pure magic that speeds code up by compiling it
# not necessary for code to run, just comment out all the @jit's
from numba import jit

from numpy.random import default_rng
rng = default_rng()
```

```python
@jit(nopython=True)
def get_b(t_history, history, tau, N=100):
    curr_time = t_history[N-1]
    norm=0
    B = 0
    for n in range(1,N):
        tmp = history[N-n]
        dt = t_history[N-n] - curr_time
        weight = np.exp(dt/tau)
        norm += weight
        tmp *= weight
        B += tmp
    
    B /= norm
    return B

# @jit(nopython=True)
# def get_bdot(t_history, history, tau, N=100):
#     curr_time = t_history[N-1]
#     norm=0
#     B1=0
#     for n in range(0,N-1):
#         tmp = history[N-n]
#         dt = t_history[N-n] - curr_time
#         weight = np.exp(dt/tau)
#         norm += weight
#         tmp *= weight
#         B1 += tmp
        

#     B1 /= norm
#     curr_time = t_history[N-2]
#     B2=0

#     for n in range(1,N-2):
#         tmp = history[N-n]
#         dt = t_history[N-n] - curr_time
#         weight = np.exp(dt/tau)
#         tmp *= weight
#         norm += weight
#         B2 -= tmp

#     B2 /= norm
    
#     return (B1-B2)/(t_history[N-1]-t_history[N-2])
```

```python
# Generates the sample data from the fake magnetometer

BUFFER_LEN=50000

# consistently, linearly spaced time points
T_clean = np.linspace(0,100,BUFFER_LEN)
# the 'original' signal
Y_clean = np.sin(T_clean)

# noise in time
t_stdev = 1e-5
# noise in y
y_stdev = 1e-3

# add some noise to simulate timer instability
T=T_clean+rng.standard_normal(BUFFER_LEN)*t_stdev
# T=T.sort()
# add some magnetometer noise
Y=np.sin(T) +rng.standard_normal(BUFFER_LEN)*y_stdev

# the true dy/dt value
true_dY = np.cos(T_clean)

plt.plot(T_clean,Y_clean)
plt.plot(T,Y)
plt.title('Generated magnetometer data')
plt.ylabel('B field')
plt.xlabel('time')
```

```python
domain = 5000
num_for_samp=30

dB = np.zeros(domain)
dB2 = np.zeros(domain)
B = np.zeros(domain)
cY = np.zeros(domain)


plt.title('Estimated dB/dt')
          
# for filter_tau in [0.001,0.01,0.1,1,10]:
for filter_tau in [3]:

    for i in range(1,domain):
        samp_Y = Y[i:i+num_for_samp]
        cY[i] = Y[i+num_for_samp-1]
        samp_T = T[i:i+num_for_samp]
        B[i] = get_b(samp_T,samp_Y, tau=filter_tau,N=num_for_samp)
        dB2[i] = (samp_Y[num_for_samp-1]-samp_Y[num_for_samp-2])/(samp_T[num_for_samp-1]-samp_T[num_for_samp-2])
        dB[i] = (B[i]-B[i-1])/(samp_T[num_for_samp-1]-samp_T[num_for_samp-2])

    plt.plot(T[:domain],dB2, label="naive dB/dt, tau=%f"%filter_tau)
    plt.plot(T[:domain],dB, label="filtered dB/dt, tau=%f"%filter_tau)
              
plt.plot(T_clean[:domain],true_dY[num_for_samp:domain+num_for_samp],'r:',label="true dB/dt")
plt.ylim([-3,3])
plt.legend(bbox_to_anchor=(1.05, 1))
```

Note that increasing the time constant makes things better, but only up to a point. After around tau=0.1 (similar to the timescale of the sampling), larger tau does not make the measured dB/dt any smoother.
